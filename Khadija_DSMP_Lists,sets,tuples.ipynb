**`Question 1:`** Class Attendance using a List

You are keeping track of student attendance using a list. Sometimes students may enter their names more than once.

*   **Step 2:** Print the list to see the attendance.

*   **Step 3:** Count how many names (including duplicates) are in the list and print the total number


```python
#We can make a hard coded list using the following line
attendance_list=["John", "Emma", "John", "Ava", "Sophia", "Emma"]
#If we want to make a list in which we enter the names one by one,we can use loop for that like following
for i in range(2):
  attendance_list.append(input("Enter the name of Student"))  #In this case we have to define the number of students first like I have defined 2
print(attendance_list)

#Now to count the number of items in a lsit we can simply use a buiilt in function
print("Total number of students in the list are ",len(attendance_list))
```

**Question 2:** Unique Attendance using a Set

---


To find out how many unique students attended the class, use a set.


*   **Step 1:** Convert the attendance_list from Question 1 into a set.
*  **Step 2:** Print the set to see the unique student names.

*   **Step 3:** Count and print the total number of unique students.



```python
#We all know that set will remove the duplicates from the list
attendance_list=["John", "Emma", "John", "Ava", "Sophia", "Emma"]
attendance_set={}
for i in range(len(attendance_list)):
  attendance_set.add(attendance_list[i])

#Now when we print,unique studets will be displayed
print(attendance_set)

#Now to count we cannot use loop as sets have no order
#we can use built in function
print("Total Number of Unique students(using built-in function) are",len(attendance_set))
```
